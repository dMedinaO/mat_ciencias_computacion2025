```python
def check_is_normal(data, alpha=0.05):
    statistics_value = stats.shapiro(data.values)
    if statistics_value[1]>alpha:
        return False
    else:
        return True
```

```python
n = len(df_data)
mean = np.mean(df_data["age_years"].values)
sd = np.std(df_data["age_years"].values, ddof=1)

t_crit = stats.t.ppf(1 - alpha/2, df=n-1)
margin = t_crit * sd / np.sqrt(n)

IC = (mean - margin, mean + margin)
IC
```
