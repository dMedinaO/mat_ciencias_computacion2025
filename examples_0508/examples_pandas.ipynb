```python
response = df["body_mass_g"]
response = response.fillna(response.mean())
data = df[['bill_length_mm', 'bill_depth_mm',
       'flipper_length_mm']]
data = data.fillna(data.mean())
```

```python
data
```

```python
pip install -U scikit-learn
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
X_train, X_val, y_train, y_val = train_test_split(data, response, random_state=42, test_size=.2)
```

```python
lr_model = LinearRegression()
lr_model.fit(X=X_train, y=y_train)
```

```python
predictions = lr_model.predict(X_val)
predictions
```

```python
r2_score(y_true=y_val, y_pred=predictions)
```

```python
data_pred = pd.DataFrame()
data_pred["predictions"] = predictions
data_pred["real_values"] = y_val
```

```python
sns.scatterplot(
    data=data_pred,
    x="predictions",
    y="real_values"
)
```
