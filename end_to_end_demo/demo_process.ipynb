```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
def make_data_visualization(
        dataset:pd.DataFrame, 
        column:str, 
        type_plot:str="box", 
        hue:str="label", 
        path_export:str=""):

    plt.clf()
    
    if type_plot == "box":
        sns.boxplot(data=dataset, x=column, hue=hue)
    elif type_plot == "hist":
        sns.histplot(data=dataset, x=column, hue=hue)
    else:
        sns.violinplot(data=dataset, x=column, hue=hue)
    
    if path_export is not None:
        plt.savefig(f"{path_export}{column}.png", dpi=300)
```

```python
data = load_breast_cancer()
X = pd.DataFrame(data.data, columns=data.feature_names)
y = pd.Series(data.target)
```

```python
X["label"] = y
X["label"].value_counts()
```

```python
for column in X.columns:
    if column != "label":
        print("Working with column: ", column)
        make_data_visualization(X, column, type_plot="violin", path_export="results/data_visualization/")
        
```

```python
from sklearn.ensemble import IsolationForest
```

```python
isolation_instance = IsolationForest(random_state=42)
isolation_instance.fit(X.drop(columns=["label"]))
X["is_isolated"] = isolation_instance.predict(X.drop(columns=["label"]))
X["is_isolated"].value_counts()
```

```python
X.groupby(by=["label", "is_isolated"]).count()
```

```python
plt.figure(figsize=(20, 20))

X_values = X.drop(columns=["label", "is_isolated"])

sns.heatmap(data=X_values.corr(method="spearman"), cmap="Blues", annot=True, fmt=".2f")
```

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
my_pipeline = Pipeline([
    ("select_features", VarianceThreshold(threshold=(.8 * (1 - .8)))),
    ("scaler", StandardScaler()),
    ("pca", PCA(n_components=2, random_state=42))
])
```

```python
my_pipeline_2 = Pipeline([
    ("select_features", VarianceThreshold(threshold=(.8 * (1 - .8)))),
    ("scaler", StandardScaler()),
    ("t_SNE", TSNE(random_state=42))
])
```

```python
my_pipeline.fit(X_values)
```

```python
pca_values = my_pipeline.transform(X_values)
```

```python
my_pipeline_2.fit(X_values)
```

```python
t_sne_values = my_pipeline_2.fit_transform(X_values)
```

```python
df_pca = pd.DataFrame(pca_values, columns=["p1", "p2"])
df_pca["label"] = y

df_tsne = pd.DataFrame(t_sne_values, columns=["p1", "p2"])
df_tsne["label"] = y
```

```python
_, axis = plt.subplots(1,2, figsize=(10,4))

sns.scatterplot(data=df_pca, x="p1", y="p2", hue="label", ax=axis[0])
sns.scatterplot(data=df_tsne, x="p1", y="p2", hue="label", ax=axis[1])

plt.tight_layout()
```

- Clustering strategy

```python
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN, AffinityPropagation
from sklearn.metrics import silhouette_score, calinski_harabasz_score, rand_score
```

```python
preparing_data_for_clustering = Pipeline([
    ("select_features", VarianceThreshold(threshold=(.8 * (1 - .8)))),
    ("scaler", StandardScaler())
])
```

```python
k_means_instance = KMeans(random_state=42, n_clusters=4)
k_means_instance.fit(preparing_data_for_clustering.fit_transform(X_values))
```

```python
silhouette_score(
    X=preparing_data_for_clustering.fit_transform(X_values), 
    labels=k_means_instance.labels_)
```

```python
agglomerative_instance = AgglomerativeClustering(n_clusters=2)
agglomerative_instance.fit(preparing_data_for_clustering.fit_transform(X_values))
```

```python
silhouette_score(
    X=preparing_data_for_clustering.fit_transform(X_values), 
    labels=agglomerative_instance.labels_)
```

```python
dbscan_instance = DBSCAN()
dbscan_instance.fit(preparing_data_for_clustering.fit_transform(X_values))
silhouette_score(
    X=preparing_data_for_clustering.fit_transform(X_values), 
    labels=dbscan_instance.labels_)
```
